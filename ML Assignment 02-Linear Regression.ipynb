```python
#1. Loading and Preprocessing (2 marks):
#● Load the California Housing dataset using the fetch_california_housing function from sklearn.
#● Convert the dataset into a pandas DataFrame for easier handling. Handle missing values (if any) and perform necessary feature scaling (e.g., standardization).
#● Explain the preprocessing steps you performed and justify why they are necessary for this dataset.
```

```python
# Import libraries
import pandas as pd
import numpy as np
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Load dataset
housing = fetch_california_housing(as_frame=True)
df = housing.frame
print(df)

# Check for missing values
print(df.isnull().sum())


```

```python
# Features and target
X = df.drop("MedHouseVal", axis=1)
y = df["MedHouseVal"]

# Feature scaling 
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_scaled=pd.DataFrame(X_scaled,columns=X.columns)
X_scaled.head()

```

```python
# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
print(X_train, X_test, y_train, y_test)
```

```python
#2. Regression Algorithm Implementation (5 marks):
#● Implement the following regression algorithms:
#○ Linear Regression
#○ Decision Tree Regressor
#○ Random Forest Regressor
#○ Gradient Boosting Regressor
#○ Support Vector Regressor (SVR)
#● For each algorithm: Provide a brief explanation of how it works. Explain why it might be suitable for this dataset.
```

```python
#LinearRegression
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X_train,y_train)
y_pred_lr = lr.predict(X_test)
```

```python
#Linear Regression model, which works by finding the "best-fit line" that minimizes the distance between your data points and the predicted values. It learns the mathematical relationship between your input features (X_train) and the target variable (y_train) by assigning weights to each feature. Once the model is "fitted," it uses that learned equation to predict outcomes (y_pred_lr) for new, unseen data (X_test).
```

```python
#DecisionTreeRegressor
from sklearn.tree import DecisionTreeRegressor

dt = DecisionTreeRegressor(random_state=42)
dt.fit(X_train, y_train)
y_pred_dt = dt.predict(X_test)
```

```python
#A Decision Tree Regressor works by repeatedly splitting your data into smaller groups based on specific "if-then" rules. It builds a tree-like structure where each branch represents a decision based on a feature, leading down to "leaf" nodes. For regression, the final prediction is usually the average value of the training samples in that specific leaf. This approach allows the model to capture complex, non-linear relationships that a simple straight line might miss.
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

def evaluate(y_test, y_pred):
    mse = mean_squared_error(y_test, y_pred)
    mae = mean_absolute_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    return mse, mae, r2


#Evaluate Models

lr_metrics = evaluate(y_test, y_pred_lr)
dt_metrics = evaluate(y_test, y_pred_dt)

print("Linear Regression", lr_metrics)
print("Decision Tree:", dt_metrics)

```

```python
#3. Model Evaluation and Comparison (2 marks):
#● Evaluate the performance of each algorithm using the following metrics:
#○ Mean Squared Error (MSE)
#○ Mean Absolute Error (MAE)
#○ R-squared Score (R2)
#● Compare the results of all models and identify: The best-performing algorithm with justification and the worst-performing algorithm with reasoning.
```

```python
results = pd.DataFrame({
    "Model": ["Linear Regression", "Decision Tree"],
    "MSE": [lr_metrics[0],dt_metrics[0]],
    "MAE": [lr_metrics[1],dt_metrics[1]],
    "R2 Score": [lr_metrics[2],dt_metrics[2]]
})
results
```

```python
#Based on the metrics, the Decision Tree is the best-performing model because it achieved the lowest error rates (MSE: 0.494, MAE: 0.453) and the highest score (0.622), indicating a better fit for the data. Conversely, Linear Regression is the worst-performing model, as it produced higher errors and a lower score (0.576), suggesting that a simple linear relationship is less effective at capturing the patterns in this dataset.
```
